```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(
    sampling_strategy='auto',  # balances minority to match majority
    k_neighbors=5,             # uses 5 nearest neighbors to interpolate
    random_state=42            # reproducibility
)

X_train_res, y_train_res = sm.fit_resample(final_df.iloc[: , :-1], final_df.iloc[: , -1].astype('int64'))
```
